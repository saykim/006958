# 샘플 데이터 불러오기

```
import csv # csv 모듈을 불러옵니다.
data_file = open('sample_data/california_housing_train.csv') #csv 파일을 엽니다.
reader = csv.reader(data_file) # csv 파일의 내용을 읽어서 리스트로 바꿔주는 csv reader를 만듭니다.

data = [] # csv에서 읽어온 데이터를 저장할 변수입니다.
for i, row in enumerate(reader): # reader에서 한 줄씩 읽어옵니다.
  if i == 0: # i가 0이면 첫번째 줄입니다. 첫번째 줄에는 헤더(컬럼명)가 있으므로 data에 넣지 않고 출력만 합니다.
    print(row)
  else : # 두번째 줄 부터는 데이터(숫자)가 들어있습니다. 각 줄의 원소들을 float(소수) 형태로 변환해서 data에 넣어줍니다.
    data.append([float(x) for x in row])

print(data[0]) # data에 저장된 첫 번째 줄을 출력합니다. 
print(data[1]) # data에 저장된 두 번째 줄을 출력합니다. 
```

# 그래프 그려보기

```
import matplotlib.pyplot as plt # 그래프를 그릴 때 필요한 matplotlib.pyplot 모듈을 plt라는 이름으로 불러옵니다.
median_income = [x[7] for x in data] # data에서 각 줄의 8번째 원소들만 모아서 median_income에 저장합니다.
median_house_value = [x[8] for x in data] # data에서 각 줄의 9번째 원소들만 모아서 median_housde_value에 저장합니다.
plt.scatter(median_income, median_house_value) # plt의 scatter 함수를 이용해 두 데이터의 산점도를 그립니다.

print(median_income)
print(median_house_value)
```

# 선형 회귀

```
from sklearn.linear_model import LinearRegression
median_income_2d = [[x] for x in median_income]
median_house_value_2d = [[x] for x in median_house_value]
regression = LinearRegression().fit(median_income_2d, median_house_value_2d)

plt.scatter(median_income, median_house_value)
x1 = min(median_income)
x2 = max(median_income)
y1 = regression.coef_[0] * x1 + regression.intercept_
y2 = regression.coef_[0] * x2 + regression.intercept_
print(x1, x2, y1, y2)
print("R square score: %.4f"%regression.score(median_income_2d, median_house_value_2d))
plt.plot([x1, x2], [y1, y2], c='r')
```

 # [어려움]
 2. 위치별(lon, lat)로 4개 (2x2) 지역으로 나누고 각각 total_rooms와 median_house_value의 관계를 분석해 보세요

```
data_NW = []
data_NE = []
data_SW = []
data_SE = []
 
longitude = [x[0] for x in data]
latitude = [x[1] for x in data]

min_longitude = min(longitude)
max_longitude = max(longitude)
mid_longitude = min_longitude + (max_longitude - min_longitude) / 2
mid_latitude = min(latitude) + (max(latitude) - min(latitude)) / 2

for row in data:
  if row[0] < mid_longitude:
    if row[1] < mid_latitude:
      data_SW.append(row)
    else:
      data_NW.append(row)
  else:
    if row[1] < mid_latitude:
      data_SE.append(row)
    else:
      data_NE.append(row)
```

```
total_rooms_NW = [x[3] / x[4] for x in data_NW] 
median_house_value_NW = [x[8] for x in data_NW]
plt.scatter(total_rooms_NW, median_house_value_NW)

total_rooms_NW_2d = [[x] for x in total_rooms_NW]
median_house_value_NW_2d = [[x] for x in median_house_value_NW]
regression_NW = LinearRegression().fit(total_rooms_NW_2d, median_house_value_NW_2d)

x1 = min(total_rooms_NW)
x2 = max(total_rooms_NW)
y1 = x1 * regression_NW.coef_[0] + regression_NW.intercept_
y2 = x2 * regression_NW.coef_[0] + regression_NW.intercept_

plt.plot([x1, x2], [y1, y2], c='r')
```
